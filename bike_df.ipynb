```python
import pandas as pd
from datetime import datetime
```

```python
filename='/content/RailsToTrails_National Count Data_week.xlsx'
xl_file = pd.ExcelFile(filename)

dfs = {sheet_name: xl_file.parse(sheet_name)
      for sheet_name in xl_file.sheet_names}

bike_df = dfs['Total Count data (31 counters)']
bike_df.head()
```

```python
#1 Solo seleccionen los datos después del 31 de julio de 2020.

filter_date = bike_df[bike_df['Week of']>'2020-07-31']
filter_date
```

```python
#2 Solo seleccionen los datos donde la columna “Change 2019-2020” es superior al 100 % (superior 1).
filter_change = bike_df[bike_df['Change 2019-2020']>1]
filter_change
```

```python
#3 ¡Únanlos! ¿Hubo alguna semana en la segunda mitad del año (después del 31 de julio de 2020) en la que el cambio fuera superior al 100 %?
df_unido = pd.merge(filter_change, filter_date, on='Week of')
df_unido
```

No hubo ninguna semana en la segunda mitad de la semana, donde el cambio sea superior a 1.
